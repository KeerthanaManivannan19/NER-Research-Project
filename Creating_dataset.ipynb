```python
import nltk
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
import contractions
```

```python
input_text = """"19th July 2025 in Hoover, Kairo Sullivan McKenzie was murdered during a car chase through Bluff Park, and the vehicle crashed near Moss Rock Preserve sarah saw that.
12th March 2024 in Helena, Isaac Roland Ortiz was kidnapped outside a movie theater, and his phone was later found near Cahaba River Bridge emily reported this.
4th May 2023 in Montgomery, Ava Danielle Rhodes was murdered while jogging in a park, and her body was recovered beside Oak Mountain Lake lily informed police.
8th October 2022 in Birmingham, Thomas Levi Barnes was kidnapped at the local school, and his bag was later discovered near Red Mountain Tunnel alex called 911.
31st January 2021 in Tuscaloosa, Naomi Faith Powell was murdered in a parking garage, and the crime scene was secured near Lake Lurleen Park george saw everything.
15th August 2020 in Mobile, Elias Matthew Carter was kidnapped while leaving the mall, and his wallet was found near Three Mile Creek anna noticed him.
23rd December 2023 in Dothan, Mia Rose Collins was murdered at a local diner, and her body was later found by Choctawhatchee River david alerted cops.
17th June 2025 in Auburn, Jason Tyler Grant was kidnapped on his way to work, and his backpack was near Chewacla State Park maya screamed loud.
6th February 2024 in Gadsden, Claire Ivy James was murdered in front of a bank, and security footage captured the suspect near Noccalula Falls Trail ethan gave it.
28th April 2022 in Florence, Logan Brent Reed was kidnapped during a field trip, and his device was last located near Tennessee River Dam grace told officers.


 """
```

```python
input_text = input_text.replace('-', ' ')
```

```python
# creating an empty list
expanded_words = []
for word in input_text.split():
  # using contractions.fix to expand the shortened words
  expanded_words.append(contractions.fix(word))

expanded_text1= ' '.join(expanded_words)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.tokenize import sent_tokenize, word_tokenize

# Tokenize sentences
sentences = sent_tokenize(expanded_text1)

# Initialize lists to store data
sentence_numbers = []
words = []

lemmas = []

# Initialize sentence number
sentence_number = 0

# Initialize WordNet lemmatizer
lemmatizer = WordNetLemmatizer()



# Iterate through sentences
for sentence in sentences:
    sentence_number += 1  # Increment sentence number for each sentence

    # Tokenize words and remove punctuation
    words_in_sentence = [word for word in word_tokenize(sentence) if word.isalnum()]



    # Extract word, POS tag, and lemmatization information
    for word in words_in_sentence:
        sentence_numbers.append(sentence_number)
        words.append(word)


        # Perform lemmatization based on POS tag

        lemma = lemmatizer.lemmatize(word)
        lemmas.append(lemma)

# Create a DataFrame
data = {
    'Sentence Number': sentence_numbers,
    'Word': words,
    'Lemmatization': lemmas
}

df = pd.DataFrame(data)
```
